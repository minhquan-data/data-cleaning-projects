```python
import pandas as pd

# Load the FIFA 21 players dataset
df = pd.read_csv("players_21.csv")

# Display the first few rows and column names
df.head(), df.columns.tolist()
```

# 1.Kiểm tra và chuẩn hóa dữ liệu cho height_cm, weight_kg

```python
# Kiểm tra và chuẩn hóa kiểu dữ liệu cho height_cm và weight_kg
height_type = df['height_cm'].dtype
weight_type = df['weight_kg'].dtype

# Kiểm tra giá trị null và kiểu dữ liệu
height_nulls = df['height_cm'].isnull().sum()
weight_nulls = df['weight_kg'].isnull().sum()
(height_type, height_nulls), (weight_type, weight_nulls)
```

# 2.Tách cột joined thành year, month, day

```python
# Chuyển đổi cột 'joined' sang định dạng datetime
df['joined'] = pd.to_datetime(df['joined'], errors='coerce')

# Tạo các cột năm, tháng, ngày từ cột 'joined'
df['joined_year'] = df['joined'].dt.year
df['joined_month'] = df['joined'].dt.month
df['joined_day'] = df['joined'].dt.day

# Kiểm tra kết quả
df[['joined', 'joined_year', 'joined_month', 'joined_day']].head()
```

# 3.Chuyển đổi cột value_eur, wage_eur, release_clause_eur sang số nguyên

```python
df['value_eur'] = pd.to_numeric(df['value_eur'], errors='coerce').fillna(0).astype(int)
df['wage_eur'] = pd.to_numeric(df['wage_eur'], errors='coerce').fillna(0).astype(int)
df['release_clause_eur'] = pd.to_numeric(df['release_clause_eur'], errors='coerce').fillna(0).astype(int)
# Kiểm tra kết quả
df[['value_eur', 'wage_eur', 'release_clause_eur']].describe()
```

# 4.Xóa ký tự xuống dòng trong player_traits và player_tags

```python
# Thay \n bằng dấu phẩy và khoảng trắng
df['player_traits'] = df['player_traits'].astype(str).str.replace('\n', ', ')
df['player_tags'] = df['player_tags'].astype(str).str.replace('\n', ', ')

# Kiểm tra kết quả mẫu
df[['player_traits', 'player_tags']].head()
```
